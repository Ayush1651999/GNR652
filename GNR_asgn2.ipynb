```python
#import the libraries and modules needed
import numpy as np
import pandas as pd
import cvxopt
import math

# Extracting Dataset from csv file into pandas dataframe
df = pd.read_csv("C:/Users/Ayush/Desktop/IITB Courses/4th sem/GNR 652/Course material/creditcard.csv")
df.dropna()
l = len(df)
m = l*0.5
m = math.floor(m)

#Selectig the trang data 
df_train = df.iloc[1:(m+1),:]
feature = df_train.iloc[:,1:30]
label = df_train.iloc[:,30]
df = df.values
df_train = df_train.values
feature = feature.values
label = label.values

#Changing the class value to -1 and +1 
for i in range(m):
    if(label[i] == 1.0):
        label[i] = 1.0
    else :
        label[i] = -1.0

#print(feature[0][28])
z = []
z_1 = []

#forming the Q array 
for i in range(m):
    for j in range(29):
        z_1.append((feature[i][j])*(label[i]))
    z.append(z_1)
    z_1 = []
    
z = np.asarray(z)

for i in range(m):
    z[i] = np.asarray(z[i])
    

Q = np.array([])
Q_1 = np.zeros(shape=(m,m))


for i in range (m):
    for j in range (m):
        product = np.dot(z[i],z[j])
        Q_1[i][j] = product 
Q = (-0.5)*Q_1
Q = Q.astype(float)
Q = matrix(Q)
# all done with Q 

#p matrix
p = [1.0]*m
p = matrix(p)
    
#G matrix
G = np.identity(m)
G = G.astype(float)
G = matrix(G)
G = G*(-1)

#h matrix
h = [0.00]*m
h = matrix(h)

#A matrix
A = matrix(label,(1,m), 'd')

# b matrix
b = matrix(0.0)

#solving the optimization problem for classification purpose
sol = cvxopt.solvers.qp(Q, p, G, h, A, b,  kktsolver='ldl')
print(sol['x'])

#w array
w = [0]*29
for i in range(m):
    u = (sol['x'][i])*(z[i])
    w = w + u
print(w)

#bias vector, given by b
for i in range(m):
    if(label[i] == 1):
        for j in range(29):
            k1 = 0
            k3 = k1 + (sol['x'][j])*w[j]
        if(k3 < k1):
            k1 = k3
b = k1
```
